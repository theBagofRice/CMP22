# Task 2

Plot $f_5$ and $f_6$ using surface plots and contour plots.

```python
from matplotlib import cm  # color map
from numpy import *
import matplotlib.pyplot as plt
```

```python
def f5(x, y):
    return sin(x+y)*tan(0.1*x)

def f6(x, y):
    return sin(5**(1/2) + x) * y
```

```python
mf = 16

# the x and y values need to be spanned on a 2D mesh
xgrid = arange(-5,5.1,.1)
ygrid = arange(-5,5.1,.1)
X, Y = meshgrid(xgrid, ygrid)
```

```python
func5 = f5(X,Y)

fig = plt.figure(figsize=(10,8))

ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, func5, cmap = cm.viridis, rstride=1, cstride=1, edgecolor='None')
ax.contourf(X, Y, func5, zdir='z', offset=-0.2, cmap=cm.magma)


plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

plt.title('Surface Plot f5(x)', fontsize=mf)

# z axis fontsize is a bit cumbersome
for t in ax.zaxis.get_major_ticks(): 
    t.label.set_fontsize(mf)

ax.set_xlabel('x value',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
ax.set_zlabel('z value',fontsize=mf)

plt.show()
```

```python
# or as a contour
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

# zdir='z' project onto x-y-plane at given offset
ax.contourf(X, Y, func5, zdir='z', offset=0, cmap=cm.coolwarm)
plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

plt.title('Contour Plot f5(x)', fontsize=mf)

for t in ax.zaxis.get_major_ticks():
    t.label.set_fontsize(mf)

ax.set_xlabel('x value',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
ax.set_zlabel('z value',fontsize=mf)
plt.show()
```

```python
func6 = f6(X,Y)

fig = plt.figure(figsize=(10,8))

ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, func6, cmap = cm.viridis, rstride=1, cstride=1, edgecolor='None')

plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

plt.title('Surface Plot f6(x)', fontsize=mf)

# z axis fontsize is a bit cumbersome
for t in ax.zaxis.get_major_ticks(): 
    t.label.set_fontsize(mf)

ax.set_xlabel('x value',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
ax.set_zlabel('z value',fontsize=mf)

plt.show()
```

```python
# or as a contour
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

# zdir='z' project onto x-y-plane at given offset
ax.contourf(X, Y, func5, zdir='z', offset=0, cmap=cm.coolwarm)
plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

plt.title('Contour Plot f6(x)', fontsize=mf)

for t in ax.zaxis.get_major_ticks(): 
    t.label.set_fontsize(mf)

ax.set_xlabel('x value',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
ax.set_zlabel('z value',fontsize=mf)
plt.show()
```
