```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import *
from scipy.interpolate import CubicSpline
```

```python
file='data_problem5_task1.dat'
data=pd.read_csv(file, sep='     ', names=['t', 'dipole moment'], engine='python')
myx=data['t']
myy=data['dipole moment']

mf=18
```

```python
nopoints = len(myx)

x = array(range(len(myx)), dtype=double) #indices of the points
```

```python
linx=arange(0,53, 0.05)
```

## part 1

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(myx, myy, label='data_problem5_task1')

ax.set_xlabel('x point',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

legend = ax.legend(loc='best', shadow=False,fontsize=mf,borderpad = 0.1, labelspacing = 0, handlelength = 0.8)
#plt.savefig('problem3_linear.pdf',dpi=300,bbox_inches='tight')
plt.show()
```

### linear interpolation

```python
lininter=interp(myx,x, myy)#, bc_type="natural")
print(len(lininter))

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(myx,lininter, marker='x', label='linear interpolation')
ax.plot(myy,marker='>')

ax.set_xlabel('x point',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

#ax.set_xlim(0, 53)

legend = ax.legend(loc='best', shadow=False,fontsize=mf,borderpad = 0.1, labelspacing = 0, handlelength = 0.8)
#plt.savefig('problem3_linear.pdf',dpi=300,bbox_inches='tight')
plt.show()
```

### cubic interpolation

The data and the interpolation are mostly similar with small differences around the peaks.

```python
csval=CubicSpline(myx, myy, bc_type="natural")

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(linx,csval(linx, 0),label='cubic interpolation')
ax.plot(myx, myy,label='real data')
#ax.plot(myy,marker='o')

ax.set_xlabel('x point',fontsize=mf)
ax.set_ylabel('y value',fontsize=mf)
plt.xticks(fontsize=mf)
plt.yticks(fontsize=mf)

#ax.set_xlim(0, 53)

legend = ax.legend(loc='best', shadow=False,fontsize=mf,borderpad = 0.1, labelspacing = 0, handlelength = 0.8)
#plt.savefig('problem3_linear.pdf',dpi=300,bbox_inches='tight')
plt.show()
```

## part 2

We can see that the cubic interpolation as well as the real data almost overlap.
