```python
from numpy import *
import matplotlib.pyplot as plt
```

### (i) one direct sequence

advantages:
- I guess pretty easy and straight forward to code

disadvantages:
- all variables are global
- kind of a mess, not well separated code

```python
lenmin = 1
lenmax = 10
delt = 20

mylen = linspace(lenmin, lenmax, delt, dtype=double)

area = zeros(len(mylen))
circ = zeros(len(mylen))

for i,my in enumerate(mylen):
    area[i] = my**2
    circ[i] = 4*my

fig, ax = plt.subplots(figsize=(8,5))

font=14

ax.plot(mylen,area,marker='x',label='area')
ax.plot(mylen,circ,marker='o',label='circumference')

ax.xaxis.set_tick_params(labelsize=font)
ax.yaxis.set_tick_params(labelsize=font)
ax.set_xlabel('length ',fontsize=font)
ax.set_ylabel('area/circumference',fontsize=font)

legend = ax.legend(loc='best', shadow=False,fontsize=font,borderpad = 0.1, labelspacing = 0, handlelength = 0.8)

plt.show()
```

### (ii) function

splitting the stuff done into two separate functions
- one for calculating the params of the square
- one for plotting the values

advantages:
- reducing the important stuff for the user to a minimum

disadvantages:
- still need to call both functions. may lead to errors, if square function wasn't executed before

```python
def square(a):
    area = zeros(len(a))
    circ = zeros(len(a))

    for i,mya in enumerate(a):
        area[i] = mya**2
        circ[i] = 4*mya
    return area, circ

def plot(length, area, circ):
    font=14

    fig, ax = plt.subplots(figsize=(8,5))

    ax.plot(length,area,marker='x',label='area')
    ax.plot(length,circ,marker='o',label='circumference')

    ax.xaxis.set_tick_params(labelsize=font)
    ax.yaxis.set_tick_params(labelsize=font)
    ax.set_xlabel('length',fontsize=font)
    ax.set_ylabel('area/circumference',fontsize=font)

    legend = ax.legend(loc='best', shadow=False,fontsize=font,borderpad = 0.1, labelspacing = 0, handlelength = 0.8)

    plt.show()
```

```python
mylen2 = linspace(lenmin,lenmax,delt, dtype=double)

area, circ = square(mylen2)
plot(mylen2, area, circ)
```

### (iii) classes

advantages:
- much easier for the user
- code easy to use

disadvantages:
- may take a bit longer to code, but not really, so maybe a bit over the top/unnecessary for smaller, not so frequent tasks

```python
class squares:
    def __init__(self, length):
        self.a = length
        self.ar = zeros(len(self.a))
        self.c = zeros(len(self.a))
        self.f = 14 #self.f
    
    def params(self):
        for i,mya in enumerate(self.a):
            self.ar[i] = mya**2
            self.c[i] = 4*mya

    def plot(self):
        
        self.params()
        fig, ax = plt.subplots(figsize=(8,5))

        ax.plot(self.a,self.ar,marker='x',label='area')
        ax.plot(self.a,self.c,marker='o',label='circumference')

        ax.xaxis.set_tick_params(labelsize=self.f)
        ax.yaxis.set_tick_params(labelsize=self.f)
        ax.set_xlabel('length',fontsize=self.f)
        ax.set_ylabel('area/circumference',fontsize=self.f)

        legend = ax.legend(loc='best', shadow=False,fontsize=self.f,borderpad = 0.1, labelspacing = 0, handlelength = 0.8)

        plt.show()
```

```python
mylen3 = linspace(lenmin,lenmax,delt, dtype=double)
mysquare = squares(mylen3)
mysquare.plot()
```
